```python
# Import packages and data files
import pandas as pd
import re

pd.options.display.max_rows = 999

s1df = pd.read_csv("table-3.csv")
s2df = pd.read_csv('table-4.csv')
s3df = pd.read_csv('table-5.csv')
s4df = pd.read_csv('table-6.csv')
s5df = pd.read_csv('table-7.csv')
s1df.head()
```

```python
#Clean data files, specifically the viewers column and repeating rows filed with plot synopsis
def cleanupdf(df):
    df['U.S. viewers(millions)'] = df['U.S. viewers(millions)'].apply(lambda x: re.sub(pattern="(\[\d+\])", repl="", string=x))
    cleaned_df = df.iloc[::2]
    return cleaned_df
```

```python
s1df = cleanupdf(s1df)
s1df
```

```python
s2df = cleanupdf(s2df)
s2df
```

```python
s3df = cleanupdf(s3df)
s3df
```

```python
# Splitting multiple episodes from the same row into individual rows by creating new row
line = pd.DataFrame({"No.overall": 99, "No. inseason": 21, 'Title': '"Bismuth (Part 2)"', "Directed by": "Joe Johnston and Jasmin Lai", "Written and storyboarded by": "Lamar Abram, Colin Howard, Jeff Liu and Katie Mitroff", "Original air date": "August 4, 2016", 'Prod.code [84]':"1031-100", "U.S. viewers(millions)": 2.15 }, index=[39])
s3df = pd.concat([s3df.iloc[:20], line, s3df.iloc[20:]]).reset_index(drop=True)
s3df
```

```python
#Correcting the information from the original row with its own independent information
s3df.iloc[[19],[0]] = 98
s3df.iloc[[19],[1]] = 20
s3df.iloc[[19],[6]] = "1031-099"
s3df.iloc[[19],[2]] = "Bismuth (Part 1)"
s3df
```

```python
#Renaming production code column for easier integration of all tables
s3df = s3df.rename(columns={'Prod.code [84]': "Prod.code"})
```

```python
s4df = cleanupdf(s4df)
s4df
```

```python
#Similar to Season 3, splitting up episodes that are in the same row
line = pd.DataFrame({"No.overall": 112, "No. inseason": 9, 'Title': '"Gem Harvest (Part 2)"', "Directed by": "Kat Morris and Ricky Cometa", "Written and storyboarded by": "Raven M. Molisee, Paul Villeco, Hilary Florido, and Jesse Zuke", "Original air date": "November 17, 2016", 'Prod.code [84]':"1040-114", "U.S. viewers(millions)": 1.39 }, index=[15])
s4df = pd.concat([s4df.iloc[:8], line, s4df.iloc[8:]]).reset_index(drop=True)
s4df
```

```python
#Correcting information for separate episodes
s4df.iloc[[7],[0]] = 111
s4df.iloc[[7],[1]] = 8
s4df.iloc[[7],[2]] = "Gem Harvest (Part 1)"
s4df.iloc[[7],[6]] = "1040-113"
s4df
```

```python
#Manually deleting repeating columns for season 5
s5df = s5df.iloc[::2]
```

```python
#Creating rows for each individual episode at the end of season 5
line1 = pd.DataFrame({"No.overall": 158, "No. inseason": 30, 'Title': '"Change Your Mind (Part 2)"', "Directed by": "Joe Johnston, Kat Morris, and Liz Artinian", "Written and storyboarded by": "Miki Brewster, Lamar Abrams, Amber Cragg, and Hillary Florido", "Original air date": "January 21, 2019", "U.S. viewers(millions)": 0.99 }, index=[29])
line2 = pd.DataFrame({"No.overall": 159, "No. inseason": 31, 'Title': '"Change Your Mind (Part 3)"', "Directed by": "Joe Johnston, Kat Morris, and Liz Artinian", "Written and storyboarded by": "Miki Brewster, Lamar Abrams, Amber Cragg, and Hillary Florido", "Original air date": "January 21, 2019", "U.S. viewers(millions)": 0.99 }, index=[30])
line3 = pd.DataFrame({"No.overall": 160, "No. inseason": 32, 'Title': '"Change Your Mind (Part 4)"', "Directed by": "Joe Johnston, Kat Morris, and Liz Artinian", "Written and storyboarded by": "Miki Brewster, Lamar Abrams, Amber Cragg, and Hillary Florido", "Original air date": "January 21, 2019", "U.S. viewers(millions)": 0.99 }, index=[31])
s5df = pd.concat([s5df.iloc[:29], line1, s5df.iloc[29:]]).reset_index(drop=True)
s5df = pd.concat([s5df.iloc[:30], line2, s5df.iloc[30:]]).reset_index(drop=True)
s5df = pd.concat([s5df.iloc[:31], line3, s5df.iloc[31:]]).reset_index(drop=True)
s5df
```
