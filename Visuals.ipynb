```python
import pandas as pd
import re
from math import pi
from datetime import datetime as dt
import time
import numpy as np
import matplotlib.pyplot as plt
from bokeh.models.ranges import Range1d
```

```python
episodedf = pd.read_csv('steveneps.csv')
episodedf
```

```python
episodedf['Original air date'] = episodedf['Original air date'].apply(lambda x: re.sub(pattern=r'\[[^()]*\]', repl="", string=x))
episodedf
```

```python
episodedf = episodedf.rename(index=str, columns={"Original air date": "date", "Title": "episode", "U.S. viewers(millions)": "Viewers_millions"})
```

```python
episodedf.to_csv('finaleps.csv')
```

```python
episodedf.head()
```

```python
from bokeh.plotting import figure, output_file, show, ColumnDataSource
from bokeh.models import DatetimeTickFormatter, Span
```

```python
# prepare some data
source = ColumnDataSource(data=episodedf)

x = episodedf['date']
y = episodedf['Viewers_millions']
episodedf['date'] = pd.to_datetime(episodedf['date'])

tooltips = [("Episode", "@episode"), ("Date", "@date"), ("Viewers", "@Viewers_millions")]

# output to static HTML file
output_file("viewers.html")

# create a new plot with a title and axis labels
p = figure(plot_width=800, plot_height=800, title="Viewers of Steven Universe", x_axis_label='Episode Release Date', y_axis_label='Viewers in Millions', x_axis_type='datetime', tooltips = tooltips)

# add a line renderer with legend and line thickness
p.line(x, y, legend="Viewers", line_width=2)

#Add spans marking the change of schedule 
schedchange1 = time.mktime(dt(2014, 3, 17, 2, 0, 0).timetuple())*1000
firstchange = Span(location=schedchange1, dimension='height', line_color='green', line_dash='dashed', line_width=1)
p.add_layout(firstchange)

schedchange2 = time.mktime(dt(2014, 5, 14, 2, 0, 0).timetuple())*1000
secondchange = Span(location=schedchange2, dimension='height', line_color='green', line_dash='dashed', line_width=1)
p.add_layout(secondchange)

schedchange3 = time.mktime(dt(2015, 3, 12, 2, 0, 0).timetuple())*1000
thirdchange = Span(location=schedchange3, dimension='height', line_color='green', line_dash='dashed', line_width=1)
p.add_layout(thirdchange)

schedchange4 = time.mktime(dt(2015, 10, 12, 2, 0, 0).timetuple())*1000
fourthchange = Span(location=schedchange4, dimension='height', line_color='green', line_dash='dashed', line_width=1)
p.add_layout(fourthchange)

schedchange5 = time.mktime(dt(2017, 2, 10, 2, 0, 0).timetuple())*1000
fifthchange = Span(location=schedchange5, dimension='height', line_color='green', line_dash='dashed', line_width=1)
p.add_layout(fifthchange)

schedchange6 = time.mktime(dt(2018, 4, 9, 2, 0, 0).timetuple())*1000
sixthchange = Span(location=schedchange6, dimension='height', line_color='green', line_dash='dashed', line_width=1)
p.add_layout(sixthchange)

p.xaxis.formatter=DatetimeTickFormatter(
        hours=["%d %B %Y"],
        days=["%d %B %Y"],
        months=["%d %B %Y"],
        years=["%d %B %Y"],
    )
p.xaxis.major_label_orientation = pi/4
output_file('myplot.html')

# show the results
show(p)
```

```python
googledf = pd.read_csv('GoogleStevenTimeline.csv')
googledf.head()
```

```python
googledf = googledf.iloc[119:]
googledf.head()
```

```python
googledf.drop(googledf.tail(5).index,inplace=True)
```

```python
googledf.reset_index(inplace=True)
```

```python
googledf = googledf.rename(index=str, columns={"index": "date", "Category: All categories": "interest"})
```

```python
googledf.head()
```

```python
list(googledf['interest'])
```

```python
x = np.array(googledf.interest)
```

```python
interest = x.astype(int)
```

```python
googledf['date'] = pd.to_datetime(googledf['date'])
x = googledf['date']
y = interest

tooltips = [("Date", "@date"), ("Interest", "@interest")]

# create a new plot with a title and axis labels
p = figure(plot_width=800, plot_height=800, title="Google Searches for Steven Universe over Time", x_axis_label='Search Date', y_axis_label='Search Interest by Percentage', x_axis_type='datetime', tooltips = tooltips)

# add a line renderer with legend and line thickness
p.line(x, y, legend="Interest", color='red', line_width=2)

p.xaxis.formatter=DatetimeTickFormatter(
        months=["%m %Y"],
        years=["%m %Y"],
    )
# output to static HTML file
p.xaxis.major_label_orientation = pi/4
output_file('googletrend.html')

# show the results
show(p)
```
