## Create graph charting Google searches over Time

```python
# Import file and clean data 

googledf = pd.read_csv('GoogleStevenTimeline.csv')
googledf.head()
```

```python
googledf = googledf.iloc[119:]
googledf.head()
```

```python
googledf.drop(googledf.tail(5).index,inplace=True)
```

```python
googledf.reset_index(inplace=True)
```

```python
googledf = googledf.rename(index=str, columns={"index": "date", "Category: All categories": "interest"})
```

```python
googledf.head()
```

```python
## Convert data to workable format
list(googledf['interest'])
```

```python
x = np.array(googledf.interest)
```

```python
interest = x.astype(int)
```

```python
# Create plot
googledf['date'] = pd.to_datetime(googledf['date'])
x = googledf['date']
y = interest

# create a new plot with a title and axis labels
p = figure(plot_width=800, plot_height=800, title="Google Searches for Steven Universe over Time Divided by Season", x_axis_label='Search Date', y_axis_label='Search Interest by Percentage', x_axis_type='datetime')

# add a line renderer with legend and line thickness
p.line(x, y, legend="Interest", color='blue', line_width=2)

#Add spans marking the change of schedule 
season1_1 = time.mktime(dt(2013, 11, 4, 2, 0, 0).timetuple())*1000
season1beg = Span(location=season1_1, dimension='height', line_color='yellow', line_dash='dashed', line_width=5)
p.add_layout(season1beg)

season1_2 = time.mktime(dt(2015, 3, 12, 2, 0, 0).timetuple())*1000
season1end = Span(location=season1_2, dimension='height', line_color='yellow', line_dash='dashed', line_width=5)
p.add_layout(season1end)

season2_1 = time.mktime(dt(2015, 3, 13, 2, 0, 0).timetuple())*1000
season2beg = Span(location=season2_1, dimension='height', line_color='purple', line_dash='dashed', line_width=5)
p.add_layout(season2beg)

season2_2 = time.mktime(dt(2016, 1, 8, 2, 0, 0).timetuple())*1000
season2end = Span(location=season2_2, dimension='height', line_color='purple', line_dash='dashed', line_width=5)
p.add_layout(season2end)

season3_1 = time.mktime(dt(2016, 5, 12, 2, 0, 0).timetuple())*1000
season3beg = Span(location=season3_1, dimension='height', line_color='red', line_dash='dashed', line_width=5)
p.add_layout(season3beg)

season3_2 = time.mktime(dt(2016, 8, 10, 2, 0, 0).timetuple())*1000
season3end = Span(location=season3_2, dimension='height', line_color='red', line_dash='dashed', line_width=5)
p.add_layout(season3end)

season4_1 = time.mktime(dt(2016, 8, 11, 2, 0, 0).timetuple())*1000
season4beg = Span(location=season4_1, dimension='height', line_color='orange', line_dash='dashed', line_width=5)
p.add_layout(season4beg)

season4_2 = time.mktime(dt(2017, 5, 11, 2, 0, 0).timetuple())*1000
season4end = Span(location=season4_2, dimension='height', line_color='orange', line_dash='dashed', line_width=5)
p.add_layout(season4end)

season5_1 = time.mktime(dt(2017, 5, 29, 2, 0, 0).timetuple())*1000
season5beg = Span(location=season5_1, dimension='height', line_color='teal', line_dash='dashed', line_width=5)
p.add_layout(season5beg)

season5_2 = time.mktime(dt(2019, 1, 21, 2, 0, 0).timetuple())*1000
season5end = Span(location=season5_2, dimension='height', line_color='teal', line_dash='dashed', line_width=5)
p.add_layout(season5end)

p.xaxis.formatter=DatetimeTickFormatter(
        months=["%m %Y"],
        years=["%m %Y"],
    )
# output to static HTML file
p.xaxis.major_label_orientation = pi/4
output_file('googletrend.html')

# show the results
show(p)
```
