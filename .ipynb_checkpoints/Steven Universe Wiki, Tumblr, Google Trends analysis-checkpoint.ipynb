We will place all of these 894,579 pages into a DataFrame, and filter them from there.

One thing we notice is the *namespaces* involved in a MediaWiki site. 

There are several namespaces, for different types of pages; the only kind of page we are concerned with at the moment is a basic content page, which sits in namespace 0. We will filter the pages to only take in namespace 0.

```python
# dump into a DataFrame:
# page title, ns, id, revision datetime, contributor, text
count, pcount, total_pages, pages_for_df = 0, 0, len(pages), []

for p in pages:
    # convert xml to badgerfish to use dict structure
    b = bf.data(p)['page']
    pcount += 1
    if b['ns']['$'] == 0:
        # build the dict to insert into the df in two pieces: there might not be text
        d = {'title': b['title']['$'],
    #         'ns': b['ns']['$'], 
             'id': b['id']['$'], 
             'timestamp': b['revision']['timestamp']['$']
            }
        if 'username' in b['revision']['contributor']:
            d['contributor'] = b['revision']['contributor']['username']['$']
            d['contributor_id'] = b['revision']['contributor']['id']['$']
        elif 'ip' in b['revision']['contributor']:
            d['contributor'] = b['revision']['contributor']['ip']['$']
            d['contributor_id'] = b['revision']['contributor']['ip']['$']

        if 'text' in b['revision'] and '$' in b['revision']['text']:  # also, @bytes > 0?
            d['text'] = b['revision']['text']['$']
        pages_for_df.append(d)
        count += 1
        if count % 100 == 0:
            print(f"total pages written to df: {count} out of {total_pages}")
    if pcount % 20000 == 0:
        print(f"total pages checked: {pcount} out of {total_pages}")

SU_wiki_pages_df_ns0 = pd.DataFrame(pages_for_df)
```

Examining these last 1414 pages to discover ways to draw the data out in a structured fashion, we notice that every episode page has a sidebar element that is written in MediaWiki markup. We'll craft a regular expression to extract all the data from this kind of sidebar, regardless what the sidebar holds.

```python
""" extracts MediaWiki sidebar data from row via regex. """
def sidebar_extraction(row):
    # 0: |   1: field name (key)   
    # 2, 3, 4: whitespace "=" whitespace   5: (value)
    regex_com = re.compile("(\|)(\w+)(\s*)=(\s*)(.*)")
    values = regex_com.findall(row)
    vals_dict = {}
    for v in values:
        vals_dict[v[1]] = v[4]
    return vals_dict
```

For the larger wiki sections with semi- to unstructured data (paragraphs of text with wiki links), we craft another regular expression to detect the section heading, extract it as a key, and extract the content of the section as the value.

```python
""" extracts data under MediaWiki headers from row via regex. """
def content_extraction(row):
    ep_big_re = "(==)(=*)(\s*)([\w\s]+)(\s*)(==)(=*)([^=]+)" 
    # slightly hacky, but Python will NOT accept numbered groups in-pattern
    # ep_big_re = "(==|===)(\s*)([\w\s]+)(\s*)(\1)([^=]+)" # SHOULD WORK but doesn't 
    ep_big_field = re.compile(ep_big_re)
    values = ep_big_field.findall(row)
    vals_dict = {}
    for v in values:
        vals_dict[v[3]] = v[7]
    return vals_dict
```

# Hypothesis Test 1 (Chi-Squared):
# #stevenuniverse Tumblr posts: 
# Do they have a different frequency when an episode airs?

We scrape Tumblr for posts with the hashtag #stevenuniverse, and check their dates against original air dates.

The question: is the proportion of posts per day "near" the airdates different from the proportion of posts per day "far" away from the airdates?

To specify the question: We define "near" a TV episode airdate, for the purposes of talking about it on social media, as the day before to 2 days after the air date. 

For example, if an episode airs on September 5, 2015, then the "near" dates are Sept 4-7. Any Tumblr post made on one of those dates will be considered "near".

First, connect to Tumblr.

```python
client = pytumblr.TumblrRestClient(
#    '<consumer_key>','<consumer_secret>','<oauth_token>','<oauth_secret>',
)
```

Examining the Tumblr API, we notice that there is no good way to "randomly" sample the posts for a given hashtag. 

Our compromise: using the client.tagged method, we set a "before" value of each month, and scrape 50 posts per month. If there are less than 50 posts in a month, then Tumblr will return posts from before that month, and we will filter out duplicates.

```python
def dt_to_unix_ts(yr,mo,day,hr,mi):
    return int(time.mktime(datetime.datetime(yr,mo,day,hr,mi).timetuple()))

# Let's pull a sample of 50 tumblr posts from before each month from Sep 2013 to the present
# and drop them all into a DataFrame
before_time = dt_to_unix_ts(2013, 9, 1, 0, 0)
months = [(2013,9),(2013,10),(2013,11),(2013,12)]
for y in [2014,2015,2016,2017,2018]:
    for m in list(range(1,13)):
        months.append((y,m))
months.extend([(2019,1),(2019,2),(2019,3),(2019,4),(2019,5)])
```

With our timing mechanism set up, we scrape.

```python
lag = 2 # minutes
request_size = 50 # sample for that month
tag_to_scrape = "stevenuniverse"
SUtumblr = [] # list of dictionaries of tumblr posts

# Make a request for a month, pause for lag minutes, make the next request.
for m in months:
    SUtumblr.extend(client.tagged(tag=tag_to_scrape, 
                                  before=dt_to_unix_ts(m[0], m[1], 1, 0, 0), 
                                  limit=request_size))
    for i in range(lag):
        time.sleep(60) # Delay for 1 minute (60 seconds).
        print(f"{m}: Paused for {i+1} of {lag} minutes.")
```

Now that we have the dates of the posts, we compute which ones are "near" and which are "far" from airdates.

```python
# The statistical test: classify the Tumblr post dates as "near" or "far" from an airing.
# "near" = 1, "far" = 0 (in near_airdates)
samples = [(x, 1 if x in near_airdates else 0) for x in SUtumblr_dates_without_times]
```

What is the total range of dates that Tumblr scraped?

```python
datetime.datetime.strptime(samples[len(samples)-1][0], "%Y-%m-%d") \
    - datetime.datetime.strptime(samples[0][0], "%Y-%m-%d")
# 2170
```

We will run a chi-squared goodness-of-fit test to see if the observed proportion of "near" posts to "far" posts can be considered due to chance or not, based on the "expected" ratios of uniformly-distributed posts (not clustered around airdates).

```python
# chi-squared homogeneity test:
# over the 2170 day period of the samples, 434 days are "near" airdates, per our definition.
# Thus, if Tumblr posts are distributed uniformly over these days, 
# approximately 434/2170 = 20% of the posts would occur on days "near" an airdate, 
# and approximately (2170-434)/2170 = 80% of the posts would occur "far" from an airdate.

# We will test the ratios of "near" and "far".

near_ratio = sum([x[1] for x in samples]) / len(samples)
far_ratio  = 1 - near_ratio
near_ratio, far_ratio
```

Technically, our hypothesis test is: 

$H_0:$ observed(near ratio, far ratio) $=$  expected(near ratio, far ratio)

$H_1:$ observed(near ratio, far ratio) $\neq$ expected(near ratio, far ratio)

This will be checked under a chi-squared test with 1 degree of freedom.

```python
df = 1 # 1 degree of freedom - only two categories for goodness of fit!
n  = 1181 # = len(samples), number of Tumblr posts checked
alpha = 0.05 # like you do

expected = [0.2, 0.8]  # len(near_airdates) / range(airdates) = 434/2170, rest = (2170-434)/2170
observed = [near_ratio, far_ratio]
```

```python
chisq, p = st.chisquare(f_obs = observed, f_exp = expected)
chisq, p
```

We clearly fail to reject the null hypothesis: the 20%/80% expected ratio is almost exactly what we observed.

```python
def visualize_chisq(chisq_stat, df, alpha):

    # initialize a matplotlib "figure"
    fig = plt.figure(figsize=(8,5))

    # get the current "axis" out of the figure
    ax = fig.gca()

    # generate points on the x axis between -4 and 4:
    xs = np.linspace(0, 10, 200)
    ys = st.chi2.pdf(xs, df)

    ax.plot(xs, ys, 'r-', lw=5, alpha=alpha, label='chi2 pdf')
    # plot the lines using matplotlib's plot function:
    ax.plot(xs, ys, linewidth=3, color='darkblue')

    # Draw two sided boundary for critical-chi-sq
    # plot a vertical line for our measured difference in rates chi-sq-statistic
    ax.axvline(chisq_stat, color='red', linestyle='--', lw=5,label='chi-sq-statistic')

    chi_sq_crit = st.chi2.ppf(1-alpha, df)
    ax.plot(xs, ys, linewidth=3, color='darkblue')
    ax.axvline(chi_sq_crit,color='green',linestyle='--',lw=4,label='crit chi-sq-value')
    ax.fill_betweenx(ys,xs,chi_sq_crit,where = xs > chi_sq_crit)
    
    ax.legend()
    plt.show()

    return None

visualize_chisq(chisq, df=1, alpha=0.05)
```

# Hypothesis Test 2 (ANOVA):
# Ratings per episode between seasons:
# Is there a difference?

Our next hypothesis test is: is there a difference in the mean viewership (as reported by Nielsen) over the seasons?

Let $\mu_i$ be the average rating per episode in season $i$.
    
$$H_0: \,\, \forall i, j \in \{1,2,3,4,5\}, \mu_i = \mu_j$$
$$H_1: \,\, \exists i, j \in \{1,2,3,4,5\}, \mu_i \neq \mu_j$$

This is an ANOVA test.

As the viewership data is in the DataFrame, we allow Statsmodel to categorize it by season and run the ANOVA test.

```python
# ANOVA on viewers_by_season; unequal sample sizes
# first, remove the "season 0" / "out-of-season" materials
episodes_df_seasons = episodes_df_multipart[episodes_df_multipart['season'] > 0]

formula = 'viewers ~ C(season)'
lm = ols(formula, episodes_df_seasons).fit()
table = sm.stats.anova_lm(lm, typ=2)
print(table)
```

We reject the null hypothesis: there is a significant difference in the means of the seasonal viewership.

We can see that Season 5's viewership is significantly lower than the others.

```python
# boxplot for means
episodes_df_seasons.boxplot('viewers', by='season', figsize=(12, 8))
```

Let's visualize the F statistic that is calculated for ANOVA.

```python
def visualize_anova(f_stat, dfn, dfd, alpha):

    # initialize a matplotlib "figure"
    fig = plt.figure(figsize=(8,5))

    # get the current "axis" out of the figure
    ax = fig.gca()

    xs = np.linspace(0, 70, 300)
    ys = st.f.pdf(xs, dfn, dfd)

    ax.plot(xs, ys, 'r-', lw=5, alpha=alpha, label='F pdf')
    # plot the lines using matplotlib's plot function:
    ax.plot(xs, ys, linewidth=3, color='darkblue')

    # Draw two sided boundary for critical-F
    # plot a vertical line for our measured difference in rates F-statistic
    ax.axvline(f_stat, color='red', linestyle='--', lw=5,label='F-statistic')

    f_crit = st.f.ppf(1-alpha, dfn, dfd)
    ax.plot(xs, ys, linewidth=3, color='darkblue')
    ax.axvline(f_crit,color='green',linestyle='--',lw=4,label='crit F-value')
    ax.fill_betweenx(ys,xs,f_crit,where = xs > f_crit)
    
    ax.legend()
    plt.show()

    return None

visualize_anova(lm.fvalue, lm.df_model, lm.df_resid, alpha=0.05)
```
